# NLP and Spacy

```python
import pandas as pd
import numpy as np
import re
import spacy
```

```python
resume = pd.read_csv ('resume_add_skills.csv', encoding = 'ISO-8859-1', usecols = [0,1,2])
resume = resume.rename(columns = {'ï»¿ID':'ID'})
resume
```

```python
lists = [35,55,110,180,290,410,460,510,570,600,620,690,730,760,790,830,870,950,1010,1100,1110,1120,1170,1210]
category = []
text = resume['Resume'].loc[0]
Resume = []
for num in lists:
    category.append(resume['Category'].loc[num])
    Resume.append(resume['Resume'].loc[num])
print(category)
```

```python
Remove_unrelated = []
text_delete_n = re.sub(r'\\n', ' ', text)
text_delete_x = re.sub(r'\\x[a-z0-9]+', ' ', text_delete_n)
for i in range(24):
    delete_n = re.sub(r'\\n', ' ', Resume[i])
    delete_x = re.sub(r'\\x[a-z0-9]+', ' ', delete_n)
    Remove_unrelated.append(delete_x)
text_delete_x
```

```python
nlp = spacy.load('en_core_web_md')
doc = nlp(text_delete_x)
similarity_1 = []
for i in range(24):
    doc_compare = nlp(Remove_unrelated[i])
    score = doc.similarity(doc_compare)
    similarity_1.append(score)
similarity_1
```

```python
d = {'label':lists,'category': category, 'score_Version_1': similarity_1} # Version 1: only delete \n and \\x\\x\\x
df = pd.DataFrame(data=d)
df
```

```python
from spacy.lang.en.stop_words import STOP_WORDS # Remove stopwords
stopwords = nlp.Defaults.stop_words
```

```python
def remove_stopwords(text):
    doc = nlp(text.lower())
    result = [token.text for token in doc if token.text not in nlp.Defaults.stop_words]
    return " ".join(result)
```

```python
text_remove_stopwords = remove_stopwords(text_delete_x) # str
doc_remove_stopwords = nlp(text_remove_stopwords)
```

```python
similarity_2 = []
for i in range(24):
    doc_compare_remove_stopwords = remove_stopwords(Remove_unrelated[i])
    doc_compare_remove_stopwords = nlp(doc_compare_remove_stopwords)
    score = doc_remove_stopwords.similarity(doc_compare_remove_stopwords)
    similarity_2.append(score)
similarity_2
```

```python
df['score_Version_2'] = similarity_2 # Remove Stopwords
df
```

```python
def lemmatization(text):
    doc = nlp(text.lower())
    result = [token.lemma_ for token in doc]
    return " ".join(result)
```

```python
text_lemmatization = lemmatization(text_remove_stopwords) # str
doc_lemmatization = nlp(text_lemmatization)
doc_lemmatization
```

```python
for entity in doc_lemmatization.ents:
    print(entity.text, entity.label_)
```

```python
def get_skills(text):
    doc = nlp(text)
    myset = []
    subset = []
    for ent in doc.ents:
        if ent.label_ == "SKILL":
            subset.append(ent.text)
    myset.append(subset)
    return subset


def unique_skills(x):
    return list(set(x))
```
