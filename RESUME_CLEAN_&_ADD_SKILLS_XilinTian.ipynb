Using Python to add the skills

```python
def score(skills, resume):
    score = 0
    for i in range(len(resume)-len(skills)):
        if resume[i:i+len(skills)]==skills:
            score = 1
    return score
```

```python
skills = ['branding','collateral','competitive analysis','content','customer journeys','data analytics','ecommerce','email marketing','events','growth','inbound marketing','influencer marketing','new market development','paid media and advertising','positioning','SEO','social media','webinars','website development','WordPress']
```

```python
for j in range(len(skills)):
    column = []
    for i in range(3677):
        skills_score = score(skills[j], resume.loc[i,'Resume'])
        column.append(skills_score)
    resume[skills[j]] = column
resume = resume.drop(columns = ['index'])
resume
```

```python
def test(r,R):
    score = True
    for i in range(len(r)):
        if r[i] != R[i]:
            score = False
    return score
```

```python
compare = []
for i in range(len(skills)):
    TF = test(resume[skills[i]], RESUME[skills[i]])
    compare.append(TF)
compare
```
