# SIMILARITY OF SKILLS


```python
import pandas as pd
import re
import plotly.express as px
```

```python
resume = pd.read_csv ('Extract skills.csv')
resume
```

```python
skill_set = resume['skills']
def change_list_to_words(text):
    a = re.sub(r'\'', '', text)
    b = re.sub(r'\[|\]', '',a)
    c = re.split(r"[,]\s*", b)
    return c
```

```python
Total_skills = []
for skillset in skill_set:
    skills = change_list_to_words(skillset)
    for skill in skills:
        Total_skills.append(skill.lower())
fig = px.histogram(
    x=Total_skills,
    labels={"x": "Skills"},
    title=f"Distribution of Skills",
).update_xaxes(categoryorder="total descending")
fig.show()
```

SÃ¸rensenâ€“Dice coefficient:
${\displaystyle DSC={\frac {2|X\cap Y|}{|X|+|Y|}}}$

```python
def DSC_score(resume, resume_compare):
    lst = [value for value in resume if value in resume_compare]
    score = 2*len(lst)/(len(resume) + len(resume_compare))
    return score
```

```python
example = ['branding','collateral','competitive analysis','content','customer journeys','data analytics','ecommerce','email marketing','events','growth','inbound marketing','influencer marketing','new market development','paid media and advertising','positioning','SEO','social media','webinars','website development','WordPress']
for k in range(len(example)):
    example[k] = example[k].lower()
similarity_score = []
for i in range(3677):
    skills = change_list_to_words(skill_set[i])
    for j in range(len(skills)):
        skills[j] = skills[j].lower()
    skills = list(set(skills))
    score = DSC_score(example, skills) 
    string = "{} -- Resume {}".format(score, i)
    similarity_score.append(string)
    

similarity_score.sort(reverse=True) # sort the list in descending list
print('The firt tenth matches for the first test data are:', similarity_score[:10])
```
